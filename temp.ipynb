```python
import cv2, os

video_path = r"C:\Users\ronitk\Downloads\WhatsApp Video 2026-02-21 at 5.20.44 PM.mp4"
out_dir = "frames"
fps_target = 2

os.makedirs(out_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_interval = max(int(round(fps / fps_target)), 1)

i = 0
saved = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    if i % frame_interval == 0:
        cv2.imwrite(os.path.join(out_dir, f"out_{saved:06d}.jpg"), frame)
        saved += 1
    i += 1

cap.release()
print("Saved frames:", saved)
```
