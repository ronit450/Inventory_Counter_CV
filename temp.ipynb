```python
import cv2, os

video_path = r"C:\Users\ronitk\Downloads\WhatsApp Video 2026-02-21 at 5.20.44 PM.mp4"
out_dir = "frames"
fps_target = 2

os.makedirs(out_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_interval = max(int(round(fps / fps_target)), 1)

i = 0
saved = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    if i % frame_interval == 0:
        cv2.imwrite(os.path.join(out_dir, f"out_{saved:06d}.jpg"), frame)
        saved += 1
    i += 1

cap.release()
print("Saved frames:", saved)
```

```python
import os
import subprocess
from pathlib import Path


def extract_frames(
    input_dir,
    output_dir,
    fps=None,                 # e.g., 1 for 1 frame/sec (None = all frames)
    image_format="png",       # png or jpg
    resize=None,              # e.g., (640, 480) or None
    quality=2                 # for jpg (2=high, 31=low)
):
    input_dir = Path(input_dir)
    output_dir = Path(output_dir)

    if not input_dir.exists():
        raise ValueError(f"Input folder does not exist: {input_dir}")

    output_dir.mkdir(parents=True, exist_ok=True)

    video_extensions = [".mp4", ".avi", ".mov", ".mkv"]

    for video_file in input_dir.iterdir():
        if video_file.suffix.lower() not in video_extensions:
            continue

        video_name = video_file.stem
        video_output_dir = output_dir / video_name
        video_output_dir.mkdir(exist_ok=True)

        output_pattern = video_output_dir / f"frame_%05d.{image_format}"

        cmd = ["ffmpeg", "-i", str(video_file)]

        # FPS filter
        vf_filters = []
        if fps:
            vf_filters.append(f"fps={fps}")

        # Resize filter
        if resize:
            width, height = resize
            vf_filters.append(f"scale={width}:{height}")

        if vf_filters:
            cmd.extend(["-vf", ",".join(vf_filters)])

        # JPG quality
        if image_format.lower() == "jpg":
            cmd.extend(["-q:v", str(quality)])

        cmd.append(str(output_pattern))

        print(f"Processing: {video_file.name}")
        subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    print("✅ Done extracting frames.")


# ----------------------------
# Example usage
# ----------------------------
if __name__ == "__main__":
    extract_frames(
        input_dir=r"/home/ronit/Downloads/videos/videos",        # your input folder
        output_dir=r"/home/ronit/Downloads/videos/frames",       # output folder
        fps=1,                    # None = all frames, or set like 1, 5, etc.
        image_format="jpg",       # png or jpg
        resize=(640, -1),         # keep aspect ratio with -1
        quality=2                 # only for jpg
    )
```
